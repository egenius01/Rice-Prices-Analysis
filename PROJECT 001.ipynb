```python
df3 = {'months':['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'],
      'imported':[239.0,244.25,270.14,298.86,316.83,336.56,338.07,370.3,365.21,388.79,410.53,410.5523],
       'local':[163.29, 172.74, 200.43, 217.94, 235.01,
        258.67, 268.27, 289.3, 284.50, 281.67, 298.68, 298.75,]
       }
```

```python
df3=pd.DataFrame(df3)
df3
```

```python
label=df3['months'][:12]
size1=df3['local'][:12]
size2=df3['imported'][:12]
```

```python
plt.bar(label,size1,color='g',width=0.8)
plt.xlabel("months of 2017")
plt.ylabel("Prices of Local Rice($)")
plt.title("Prices of Locally made Rice, Every Sixtheent Day of Every Month in 2017\n\n")
```

```python
plt.bar(label,size2,color='r',width=0.8)
plt.bar(label,size1,color='g',width=0.8)
legend=['Imported','Local']
plt.legend(legend)
plt.xlabel("months of 2017")
plt.ylabel("Prices of Imported Rice ($)")
plt.title("Prices of Imported Rice, Every Sixtheent Day of Every Month in 2017\n\n")
```
